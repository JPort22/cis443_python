```python
#Program 2
#Grading ID: P9941
#CIS 443-01

#1st function = input_test_scores -> prompts user for how many students and how many tests for each student
#then uses a loop to input this data into a 2-D list of student test scores
#will return this 2-D list of student test scores

#2nd function = summarize_test_scores -> will accept the 2-D list of test scores as its parameter and calculate the mean test score
#and associated letter grade for each student
#function should create and display a bar chart of the letter grade frequencies

%matplotlib inline
import matplotlib.pyplot as plt #imports matplotlib library to enable the graphics for the bar chart
import numpy as np
import seaborn as sns
import statistics # could also use sum len to calculate mean of students test scores to get their corresponding letter grade

def input_test_scores(): #function takes in 2 parameters one for the student amount and one for test amount
    input_student = int(input('How many students?')) #receive number of students from user
    input_test = int(input('How many tests?')) #receive number of tests from user
    student_test_scores = [] #empty 2D list
    for i in range(input_student): #for loop steps into the range of the number of students
        #student_test_scores.append([]) #number of rows that are then appended to 2D list student_test_scores
        row_student = [] #list that represents row of student grades
        print(f'Enter {input_test} scores for student {i + 1}') #prints the number of tests given alongwith counting off number of students
        for j in range(input_test): #for loop steps through 
            #student_test_scores[i].append(0)
            row_student.append(int(input('Enter score:'))) #this list appends entered scores into the student row
        student_test_scores.append(row_student) #appends each row from 1 students test scores into 2D list
    return student_test_scores #returns the 2D list

def grade_chart(mean_score): #function connects mean_score to a grade letter for student
    grade = 'F' #defaults to F
    
    if mean_score >= 90: #a grade letter given
        grade = 'A'
    elif mean_score >= 80: #b grade letter given
        grade = 'B'
    elif mean_score >= 70: #c grade letter given
        grade ='C'
    elif mean_score >= 60: #d grade letter given
        grade = 'D'
    
    return grade

def summarize_test_scores(student_test_scores):#function takes in student_test_scores
    grade_list = [] #empty list to put the grades into
    for row in student_test_scores:#for each row of students in student_test_scores
        student_mean = statistics.mean(row) #calculate student_mean
        print(student_mean)
        grade = grade_chart(student_mean)#grade assigned to student_mean from grade_chart
        grade_list.append(grade)#each grade from each student added to grade_list
    print(grade_list)
    values, frequencies = np.unique(grade_list, return_counts=True) #be careful about letter grades that may be 0 students attached
    title = 'Letter Grades Bar Graph' #title of bar graph
    sns.set_style('whitegrid')# style of bar graph
    axes = sns.barplot(x=values, y=frequencies, palette='bright') #creates and displays bar graph
    axes.set_title(title)
    axes.set(xlabel='Letter Grade', ylabel='Frequency')# sets labels for both y and x axis
    axes.set_ylim(top=max(frequencies) * 1.10)#sets shaping/sizing for bar graph
    for bar, frequency in zip(axes.patches, frequencies):# for loop that steps through frequency of grades occurring
        text_x = bar.get_x() + bar.get_width() / 2.0  
        text_y = bar.get_height() 
        text = f'{frequency:,}\n{frequency / len(grade_list):.3%}'
        axes.text(text_x, text_y, text, fontsize=11, ha='center', va='bottom')

test_scores = input_test_scores()
#test_scores = [[67, 54, 55], [95, 97, 100], [87, 82, 84], [73, 76, 70], [90, 95, 91]]
summarize_test_scores(test_scores)
```
