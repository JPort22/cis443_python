```python
#Grading ID: P9941
#CIS 443-01, Program 1 - due 9/20

#Purpose: Gain user input to calculate the length of pi decimal

pi_sum = 0 #pi summation starting point
num = 4 #numerator constant
denom = 1 #denominator starting point
#user input text box/field
users_term = int(input('Enter the number of terms you wish to see:'))
#for loop for going through range of user inputted term and taking the current
#term for pi estimation
for current_term in range(users_term + 1):
    #current_term -= 2
    if current_term % 2 == 0: #if current term led to a remainder of 0
        pi_sum += num / denom #then positive multiplier for division
        #denom += 2
    else: #if not leading to a remainder of 0
        pi_sum -= num / denom #then negative multiplier for division
    denom += 2 #denominator incrementing by 2
    print(f'{current_term:>8}{pi_sum:>10.6f}')#output printing style
```
